```
# I need to equate the strings "rock", "paper", "scissors", "lizard", "Spock" 
# to numbers:
# 0 - rock
# 1 - Spock
# 2 - paper
# 3 - lizard
# 4 - scissors

# I need to import the following package to python to generate random
# numbers

import random

# I need to names to numbers
def number_to_name(number):
    if number == 0:
        return "rock"
    elif number == 1:
        return "Spock"
    elif number == 2:
        return "paper"
    elif number == 3:
        return "lizard"
    elif number == 4:
        return "scissors"
    else:
        return "Error"

# I need to assign numbers to names
def name_to_number(name):
    if name == "rock":
        return 0
    elif name == "Spock":
        return 1
    elif name == "paper":
        return 2
    elif name == "lizard":
        return 3
    elif name == "scissors":
        return 4
    else:
        print (name + " is not a character in the RPSLS_game")

# function which selects the winner
def RPSLS_game(name): 
    print("Let's play!")
    player_number = name_to_number(name);
    print ("Player chose " + name)
    #  I need to generate randomly the choice of the computer
    #  I created a list of possible integer values
    num_list= [0, 1, 2, 3, 4]
    # Use random.choice to generate random value
    comp_number = random.choice(num_list)
    print ("Computer chose " + number_to_name(comp_number))

    # compute difference of player_number and comp_number modulo five
    score = (player_number - comp_number) % 5
    if (score == 0):
        print ("Player and computer tie!");
    elif (score <= 2):
        print ("Player wins!")
    else:
        print ("Computer wins!")
    print

RPSLS_game("rock")
RPSLS_game("scissors")
RPSLS_game("lizard")


```
